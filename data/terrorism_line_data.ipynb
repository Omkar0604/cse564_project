```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('terrorism_data_clean.csv', index_col=False)
```

```python
df_lp = df[['iyear', 'country_txt', 'nkilled', 'nwounded', 'start_year']]
df_lp.head()
```

```python
country_name = 'Mexico'
year = 1980
df_lp = df_lp[df_lp['country_txt'] == country_name]
df_lp = df_lp[df_lp['start_year'] == year]
df_lp = df_lp.drop(['start_year', 'country_txt'], axis = 1)
```

```python
df_lp_agg = df_lp.groupby(['iyear']).sum()
df_lp_agg = df_lp_agg.reset_index()
df_lp_agg = df_lp_agg.rename(columns={'iyear': 'Year', 'nkilled': 'Number Killed', 'nwounded': 'Number Wounded'})
df_lp_agg
```
