```python
df_2017_clean = df_2017.copy()
```

```python
# np.isnan(list(df['Cost of an energy sufficient diet [CoCA]'].map(lambda l: l))[0])
df_2017_clean = df_2017_clean[df_2017_clean['Country Name'] != 'World']
df_2017_clean = df_2017_clean[df.columns[6:]].astype(float)
```

```python
for index in range(len(df_2017_clean.columns)):
    col = df_2017_clean.columns[index]
    col_min = df_2017_clean[col].min()
    col_max = df_2017_clean[col].max() / 10
    df_2017_clean[col] = df_2017_clean[col].map(lambda l: l if not np.isnan(l) else np.random.choice([col_min, col_max]))
```
